2. Considering the GDP and population size, plot the material consumption intensity, per capita DMC and per capita material footprint (i.e., RMC) of Switzerland, China, and the United States from 2000 to 2015. 

```python
# import libraries
import matplotlib.pyplot as plt
```

```python
# import GDP data
df_GDP = pd.read_excel('OECD_Dataset_GDP_2000_2015.xlsx',
                    header=[1], 
                    index_col=[0],
                    skiprows=[1,2,4,5,22,23,24],
                    ) 
df_GDP.head() 
```

```python
# material intensity

fig = plt.figure(figsize = (12,4))

for country in ["China (People's Republic of)",'Switzerland','United States']:
    x = range(2000,2016,1)
    y = data.loc[:,('Domestic Material Consumption',country)] / df_GDP.loc[:,country]
    
    plt.plot(x,y,label=country)
    
    plt.xlabel('Year')
    plt.ylabel('tonnes/USD')
    
    plt.legend()
    plt.grid()
    
plt.show()
```
