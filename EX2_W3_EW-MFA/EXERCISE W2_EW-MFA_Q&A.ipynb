2. Considering the GDP and population size, plot the material consumption intensity, per capita DMC and per capita material footprint (i.e., RMC) of Switzerland, China, and the United States from 2000 to 2015. 

```python
# import libraries
import matplotlib.pyplot as plt
```

```python
# import GDP data
df_GDP = pd.read_excel('OECD_Dataset_GDP_2000_2015.xlsx',
                    header=[1], 
                    index_col=[0],
                    skiprows=[1,2,4,5,22,23,24],
                    ) 
df_GDP.head() 
```

```python
# material intensity

fig = plt.figure(figsize = (12,4))

for country in ["China (People's Republic of)",'Switzerland','United States']:
    x = range(2000,2016,1)
    y = data.loc[:,('Domestic Material Consumption',country)] / df_GDP.loc[:,country] 
    plt.plot(x,y,label=country)
    
    plt.xlabel('Year')
    plt.ylabel('tonnes/USD')
    
    plt.legend()
    plt.grid()
    
plt.show()
```

```python
# per capita DMC

fig = plt.figure(figsize = (12,4))

for country in ["China (People's Republic of)",'Switzerland','United States']:
    x = range(2000,2016,1)
    y = data.loc[:,('Domestic material consumption per capita',country)]
    plt.plot(x,y,label=country)
    
    plt.xlabel('Year')
    plt.ylabel('Kilograms per capita, Thousands')
    
    plt.legend()
    
    plt.grid()
    
plt.show()
```

```python
# per capita material footprint (i.e., raw material consumption - RMC)

fig = plt.figure(figsize = (12,4))

for country in ["China (People's Republic of)",'Switzerland','United States']:
    x = range(2000,2016,1)
    y = data.loc[:,('Material footprint per capita',country)]
    plt.plot(x,y,label=country)
    
    plt.xlabel('Year')
    plt.ylabel('Kilograms per capita, Thousands')
    
    plt.legend()
    
    plt.grid()
    
plt.show()
```

Establish the economy-wide MFAs for France, the United Kingdom, and Spain (2015). Consider only the following flows into and out of the economy: domestic extraction of materials, imports, domestic processed outputs, and exports (see Figure 1). Flows within the economy can be ignored.

```python
# Display external images using IPython.display.Image
from IPython.display import Image 
Image(filename='EX2_EW-MFA scheme.PNG', width=700) 
```

```python
# import DPO data
df_DPO = pd.read_excel('EUROSTAT_Dataset_DPO_2014_2016.xlsx',
                    header=[1,2], 
                    index_col=[0],
                    skiprows=range(1,8),
                    ) 
df_DPO
```

```python
# Extract the indicators from the xlsx datasets

countries = ['France', 'United Kingdom', 'Spain']
year = 2015

# domestic extraction of materials
DEU = {}
imports = {}
exports = {}
DPO = {}

for country in countries:
    DEU[country] = data.loc[year,('Domestic extraction used',country)]
    imports[country] = data.loc[year,('Imports',country)]
    exports[country] = data.loc[year,('Exports',country)]
    DPO[country] = df_DPO.loc[country,('Thousand tonnes',str(year))] # Here, year is a string and not an integer

```

```python
DPO
```

```python
# DMI = DEU + IMP
DMI = {}
for country in countries:
    DMI[country] = DEU[country] + imports[country]
    print(f'DMI of {country}: {DMI[country]} Tonnes, Millions')
```

```python
# DMC = DMI - EXP
DMC = {}
for country in countries:
    DMC[country] = DMI[country] - exports[country]
    print(f'DMC of {country}: {DMC[country]} Tonnes, Millions')
```

```python
# Physical trade balance (PTB): IMP -EXP
PTB = {}
for country in countries:
    PTB[country] = imports[country] - exports[country]
    print(f'PTB of {country}: {PTB[country]} Tonnes, Millions')
```

```python
# resource productivity: GDP/DMI (million USD/million tonnes)
res_productivity = {}
for country in countries:
    res_productivity[country] = df_GDP.loc[year,country] / DMI[country]
    print(f'The resource productivity of {country}: {res_productivity[country]} [USD/ tonnes]')
```
